```python
import numpy as np
import pandas as pd
import sklearn as sk
import seaborn as sns
import matplotlib.pyplot as plt
import statistics
from sklearn.linear_model import LogisticRegression
from imblearn.under_sampling import RandomUnderSampler
from sklearn.svm import LinearSVC
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import f1_score
```

```python
import warnings

def isnat(x):
    return x.astype('i8') == np.datetime64('NaT').astype('i8')

def clean_white_space(data):
    
    for i in range(len(data)):
        ind = 0 
        for g in data.iloc[i] :
            try:
                g = g.split()
                if g == []:
                    data.iloc[i,ind]= np.nan
                ind = ind +1
            except:
                ind = ind +1
                pass

    return (data)

def isfloat(value):
    try:
        float(value)
        return True
    
    except ValueError:
        return False

def check_types(new_data):
    
    marks = {}.fromkeys(['numeric', 'string','null','boolean','datetime'], 0)

    for index in range(len(new_data)):
        
        row = new_data[index]
        
        if ((isfloat(row)) &  (pd.isna(row)is False)):
             row=float(row)
             new_data[index]=row
             marks['numeric'] += 1
                          
        
        
        if isinstance(row,str):
                                      
            if (((row.lower() == 'true') | (row.lower() == 'false'))is False):
                    try:
                        new_data[index]=pd.to_datetime(row)
                        if isinstance(row, pd._libs.tslibs.nattype.NaTType):
                            print('row')
                            new_data[index] = np.nan()
                        marks['datetime'] += 1
                
                    except:
                          marks['string'] += 1
                     
            elif (row.lower() == 'true') | (row.lower() == 'false'):
                     marks['boolean'] += 1                  
                
        
        
        if pd.isna(row):
                marks['null'] += 1            
                print("there is null at line : " , index, )       
        
            
    print(marks)
    return(marks)


def DataCleaning(data):
    # בדיקה אם יש white space
    data = clean_white_space(data)
    # בדיקה אם יש עמודה רק של ערכי null והורדה של אחת כזו
    data = data.drop_duplicates()
    nunique = data.nunique()
    cols_to_drop = nunique[nunique == 1].index
    data = data.drop(cols_to_drop, axis=1)
    data=data.reset_index()
    data=data.drop('index',axis = 1)

    num_to_impute = []
    not_drop = []
    j=0
    
    for i in data:
        
        print("Column num:",j,":",i,":")
        new_data = data.loc[:,i]
        warnings.filterwarnings('ignore')
        marks = check_types(new_data)
        warnings.filterwarnings('ignore')
        max_key = max(marks, key=marks.get)
        warnings.filterwarnings('ignore')

        if max_key == 'numeric':
            data.loc[:,i]=pd.to_numeric(data.loc[:,i], errors='coerce')

            if new_data.isnull().mean() >= 0.8:
                data=data.drop(i , axis = 1)

        if max_key == 'datetime':
            data.loc[:,i]=pd.to_datetime(data.loc[:,i], errors='coerce')
            
            if (1-(new_data.value_counts().sum()/len(new_data))) >= 0.8:
                data=data.drop(i , axis = 1)
            
        elif   max_key == 'string': 
            data.loc[:,i].to_string()
        
        elif  max_key == 'boolean':
            data.loc[:,i]=data.loc[:,i].astype('bool')

            if new_data.isnull().mean() >= 0.8:
                data=data.drop(i , axis = 1)
            
        

        j+=1
        print("Its a" , max_key,"column")
        print()
        print("------------------------------------------------------------")
        print()
        
    return(data)


```

```python
def plot(dic,k,model):
    sns.set()
    x=[]
    mean=[]
    std=[]
    mean_plus=[]
    mean_minos=[]
    for k in k:   
        x.append(k)
        mean.append(dic[k]['mean'])
        std.append(dic[k]['std'])
        mean_plus.append(dic[k]['mean']+dic[k]['std'])
        mean_minos.append(dic[k]['mean']-dic[k]['std'])
    plt.figure(figsize=(16, 6))
    plt.plot(x, mean, 'b-', label='mean')
    plt.fill_between(x, mean_minos, mean_plus, color='b', alpha=0.2)
    plt.legend()
    plt.title("performance of {} model ".format(model))
    plt.show()
    

```

# knn

```python
dic={}
sc=StandardScaler()
k=list(map(int, np.linspace(1, 20, num=20)))
for c in k:
    lst=[] 
    for i in range(1000):
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=i)
        X_train = sc.fit_transform(X_train)
        X_test = sc.transform(X_test)
        knn = KNeighborsClassifier(n_neighbors = c)
        knn.fit(X_train, y_train)
        pred = knn.predict(X_test)
        lst.append(f1_score(y_test, pred))
    mean = statistics.mean(lst)
    std=np.std(lst)
    dic[c]={'mean': mean,'std':std}

plot(dic,k,'knn')
dic
```

# LinearSVC


```python
from sklearn.svm import LinearSVC
sc=StandardScaler()
k=np.linspace(0.0001, 2, num=21)
dic={}
for c in k:    
    lst=[]
    for i in range(1000):
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=i)
        X_train = sc.fit_transform(X_train)
        X_test = sc.transform(X_test)
        svm = LinearSVC(C=c).fit(X_train, y_train)
        pred = svm.predict(X_test)
        lst.append(f1_score(y_test, pred))
    mean = statistics.mean(lst)
    std=np.std(lst)
    dic[c]={'mean': mean,'std':std}
plot(dic,k,'LinearSVC')
dic
```

```python
m=[2,3,4,5]
dic={}
sc=StandardScaler()
for k in m:
    lst=[]
    for i in range(1000):
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=i)
        X_train = sc.fit_transform(X_train)
        X_test = sc.transform(X_test)
        svm = SVC(kernel='poly', degree=k).fit(X_train, y_train)
        pred = svm.predict(X_test)
        lst.append(f1_score(y_test, pred))
        
    mean = statistics.mean(lst)
    std=np.std(lst)
    dic[k]={'mean': mean,'std':std}
plot(dic,m,'PolySvc')
dic
max_key = 0
j=0
for i in m:
    tmp = dic[i]['mean']
    if (tmp > max_key):
        j=i
        max_key = tmp
```
